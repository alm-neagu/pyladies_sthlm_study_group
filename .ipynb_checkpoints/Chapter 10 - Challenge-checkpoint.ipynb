```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the data (2020 Kaggle Machine Learning & Data Science Survey)
data_raw = pd.read_csv('datasets/own/kaggle_survey_2020_responses.csv',
                       usecols=[1, 2, 3, 4],
                       names=['Age', 'Gender', 'Country', 'Education'],
                       skiprows=2)
data_raw.head()
```

```python
# Age distribution analysis
fig, ax = plt.subplots(1,1, figsize=(15, 6))
age = data_raw['Age'].value_counts().sort_index()

# Plot a barchart & some formatting
ax.bar(age.index, age, width=0.55, edgecolor='darkgray', color='#d4dddd', linewidth=0.7)
ax.set_ylim(0, 4500)
fig.text(0.1, 0.95, 'Age Distribution on Kaggle - Survey 2020', fontsize=15, 
         fontweight='bold', fontfamily='serif')

# Add barchart values
for i in age.index:
    ax.annotate(f"{age[i]}", 
                   xy=(i, age[i] + 100),
                   va = 'center', ha='center',fontweight='light', fontfamily='serif',
                   color='#4a4a4a')

# Show the plot
plt.show()
```

```python
# Age & gender distribution
age_gender = data_raw.groupby(['Gender'])['Age'].value_counts().unstack().sort_index()
man = age_gender.loc['Man'].astype('int')
woman = -age_gender.loc['Woman'].astype('int')

# Plot barchart with legend
fig, ax = plt.subplots(1,1, figsize=(15, 6))
ax.bar(man.index, man, width=0.55, color='#004c70', alpha=0.8, label='Male')
ax.bar(woman.index, woman, width=0.55, color='#990000', alpha=0.8, label='Female')
ax.set_ylim(-1200, 3500)

# Add barchart values
for i in man.index:
    ax.annotate(f"{man[i]}", 
                   xy=(i, man[i] + 100),
                   va = 'center', ha='center',fontweight='light', fontfamily='serif',
                   color='#4a4a4a')
    
for i in woman.index:
    ax.annotate(f"{-woman[i]}", 
                   xy=(i, woman[i] - 100),
                   va = 'center', ha='center',fontweight='light', fontfamily='serif',
                   color='#4a4a4a') 

# Some formatting
ax.set_yticks([])    
ax.legend()
fig.text(0.16, 0.95, 'Age / Gender Distribution', fontsize=15, 
         fontweight='bold', fontfamily='serif')
    
plt.show()
```
